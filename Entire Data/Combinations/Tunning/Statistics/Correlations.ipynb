```python
# Importing the libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from dython.nominal import associations
import numpy as np


# Set working directory
months = ['june', 'july', 'august', 'september', 'october']
# Set working directory
data_path = "/Users/mau/Library/CloudStorage/Dropbox/Mac/Documents/Dissertation/Chapter 2/Entire_Data/Statistics/"
```

```python
datasets = {}

for month_file in months:
        name = 'dtf' +'_'+ month_file
        dataset_path = f"{data_path}/data_for_stats_{month_file}.parquet"
        # Importing the dataset and keeping only selected columns
        dataset = pd.read_parquet(dataset_path)
        # Transform datetime to seconds
        # Convert ave_time_per_machine to seconds
        datasets[name] = dataset

# # Extract June dataset
# dtf_June = datasets['dtf_June']

# # Extract August dataset
# dtf_August = datasets['dtf_August']
# # Extract September dataset
# dtf_September = datasets['dtf_September']
# # Extract October dataset
# dtf_October = datasets['dtf_October']
```

```python
def pearson_correlation(data, classification):
    dtf = datasets[data]
    # Filter for 'top_10' classification
    dtf_top_10 = dtf[dtf['clasification'] == classification]

    # Calculate the profit to bet ratio column 'p/b'
    dtf_top_10['p/b'] = dtf_top_10['profit'] / dtf_top_10['wageredamt']

    # Remove infinite (inf) values by checking for both NaN and inf
    dtf_top_10 = dtf_top_10.replace([np.inf, -np.inf], np.nan).dropna()

    # Estimate and generate a Pearson's R association matrix
    pearson_r = associations(dtf_top_10[['playercashableamt', 'wageredamt', 'theoreticalpaybackpercent', 'age', 'profit', 'slotdenomination', 'maxbet', 'p/b']])
    # pearson_r_matrix = pearson_r['corr'].round(2)
```
