```python
# Filter Columns
filter = ['session_time', 'gender', 'age_gen', 'day', 'timeofday', 'first_outcome',
        'first_wager','first_p/b', 'last_outcome', 'last_wager', 'last_p/b',
        'beginning_amt', 'ending_amt', 'ending_balance', 'ave_slotdenom', 
        'std_slotdenom', 'min_slotdenom', 'max_slotdenom', 'ave_theo_payback',
        'min_theo_payback', 'max_theo_payback', 'ave_wageramt', 'std_wageramt',
        'min_wager', 'max_wager', 'ave_p/b', 'std_p/b', 'max_p/b', 'max_profit', 'depletion_slope', 
        '#inc_slotdenom', '#dec_slotdenom', '#inc_maxbet', '#dec_maxbet',
        'w/min', 'l/min', 'nh/min', 'd/min', '2ws_profit', '2ws_wgramt','2ws/min', 
        '3ws_profit', '3ws_wgramt', '3ws/min', '4ws_profit', '4ws_wgramt', '4ws/min', 
        'w/g', 'l/g', 'nh/g', 'd/g', 'ave_time_per_gamble',
        'min_time_per_gamble', 'machines_changes', 'unique_machines', 'ave_time_per_machine', 
        'total_duration', 'total_gambles', 'classification']

# Taking out ave_time_per_machine and max_time_per_gamble

datasets = {}

for month_file in months:
        name = 'dtf' + month_file[1:]
        dataset_path = f"{data_path}/{month_file}/Ending Balances/Per_Player/top_vs_ntop_player_ending_balance_by_session.parquet"
        # Importing the dataset and keeping only selected columns
        dataset = pd.read_parquet(dataset_path, columns=filter)
        # Transform datetime to seconds
        # Convert ave_time_per_machine to seconds
        dataset['ave_time_per_machine'] = dataset['ave_time_per_machine'].dt.total_seconds()
        dataset['total_duration'] = dataset['total_duration'].dt.total_seconds()
        datasets[name] = dataset

# # Extract June dataset
# dtf_June = datasets['dtf_June']
# # Extract July dataset
# dtf_July = datasets['dtf_July']
# # Extract August dataset
# dtf_August = datasets['dtf_August']
# # Extract September dataset
# dtf_September = datasets['dtf_September']
# # Extract October dataset
# dtf_October = datasets['dtf_October']
```

```python
import pandas as pd

# Create an empty DataFrame to store the statistics
result_df = pd.DataFrame(columns=["Month", "Category", "Count", "Mean Gain/Loss", "Max Gain/Loss", "Mean Wager", "Mean 'std_wageramt'", "Mean 'std_p/b'"])

# Iterate through your datasets
for i in datasets:
    data = datasets[i]
    data = data[data['session_time'] == 1]
    # Count only top_10 players
    top_10 = data[data['classification'] == 'top_10']
    # Count only non_top_10 players
    non_top_10 = data[data['classification'] == 'Ntop']
    # Separate players with positive ending balance
    top_10_pos = top_10[top_10['ending_balance'] > 0]
    non_top_10_pos = non_top_10[non_top_10['ending_balance'] > 0]
    # Separate players with negative ending balance
    top_10_neg = top_10[top_10['ending_balance'] < 0]
    non_top_10_neg = non_top_10[non_top_10['ending_balance'] < 0]

    # Calculate the statistics
    top_10_pos_count = len(top_10_pos)
    non_top_10_pos_count = len(non_top_10_pos)
    top_10_neg_count = len(top_10_neg)
    non_top_10_neg_count = len(non_top_10_neg)

    # Eding balances
    top_10_pos_mean_gain = top_10_pos['ending_balance'].mean()
    non_top_10_pos_mean_gain = non_top_10_pos['ending_balance'].mean()
    top_10_neg_mean_loss = top_10_neg['ending_balance'].mean()
    non_top_10_neg_mean_loss = non_top_10_neg['ending_balance'].mean()

    # Max gains and losses
    top_10_pos_max_gain = top_10_pos['ending_balance'].max()
    non_top_10_pos_max_gain = non_top_10_pos['ending_balance'].max()
    top_10_neg_max_loss = top_10_neg['ending_balance'].min()
    non_top_10_neg_max_loss = non_top_10_neg['ending_balance'].min()

    # Average 'std_p/b'
    top_10_pos_mean_std_pb = top_10_pos['std_p/b'].mean()
    non_top_10_pos_mean_std_pb = non_top_10_pos['std_p/b'].mean()
    top_10_neg_mean_std_pb = top_10_neg['std_p/b'].mean()
    non_top_10_neg_mean_std_pb = non_top_10_neg['std_p/b'].mean()

    # Average wager amount
    top_10_pos_mean_wager = top_10_pos['ave_wageramt'].mean()
    non_top_10_pos_mean_wager = non_top_10_pos['ave_wageramt'].mean()
    top_10_neg_mean_wager = top_10_neg['ave_wageramt'].mean()
    non_top_10_neg_mean_wager = non_top_10_neg['ave_wageramt'].mean()

    # Average std_wageramt
    top_10_pos_mean_std_wager = top_10_pos['std_wageramt'].mean()
    non_top_10_pos_mean_std_wager = non_top_10_pos['std_wageramt'].mean()
    top_10_neg_mean_std_wager = top_10_neg['std_wageramt'].mean()
    non_top_10_neg_mean_std_wager = non_top_10_neg['std_wageramt'].mean()

    # Append the statistics to the DataFrame
    result_df = result_df.append({"Month": i[4:], "Category": "Top 10 Positive", "Count": top_10_pos_count, "Mean Gain/Loss": top_10_pos_mean_gain,
                                  "Mean 'std_p/b'": top_10_pos_mean_std_pb, "Max Gain/Loss": top_10_pos_max_gain,
                                  "Mean Wager": top_10_pos_mean_wager, "Mean 'std_wageramt'": top_10_pos_mean_std_wager}, ignore_index=True)
    result_df = result_df.append({"Month": i[4:], "Category": "Top 10 Negative", "Count": top_10_neg_count, "Mean Gain/Loss": top_10_neg_mean_loss,
                                "Mean 'std_p/b'": top_10_neg_mean_std_pb, "Max Gain/Loss": top_10_neg_max_loss,
                                "Mean Wager": top_10_neg_mean_wager, "Mean 'std_wageramt'": top_10_neg_mean_std_wager}, ignore_index=True)
    result_df = result_df.append({"Month": i[4:], "Category": "Non-Top Positive", "Count": non_top_10_pos_count, "Mean Gain/Loss": non_top_10_pos_mean_gain,
                                  "Mean 'std_p/b'": non_top_10_pos_mean_std_pb, "Max Gain/Loss": non_top_10_pos_max_gain,
                                  "Mean Wager": non_top_10_pos_mean_wager, "Mean 'std_wageramt'": non_top_10_pos_mean_std_wager}, ignore_index=True)
    result_df = result_df.append({"Month": i[4:], "Category": "Non-Top Negative", "Count": non_top_10_neg_count, "Mean Gain/Loss": non_top_10_neg_mean_loss,
                                  "Mean 'std_p/b'": non_top_10_neg_mean_std_pb, "Max Gain/Loss": non_top_10_neg_max_loss,
                                  "Mean Wager": non_top_10_neg_mean_wager, "Mean 'std_wageramt'": non_top_10_neg_mean_std_wager}, ignore_index=True)

# Set both "Month" and "Category" as the index
result_df = result_df.set_index(["Month", "Category"])

```
