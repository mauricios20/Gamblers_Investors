## June

```python
training = ['6_October', '3_July', '4_August', '5_September']
testing = ['2_June']
best_params = {'C': 10, 'kernel': 'rbf'}

june_test, jn_y_test, jn_y_pred, jn_x_test , jn_df, jn_classifier, ct, jn_x_train, jn_y_train = cf.run_svm_combine(months, data_path, filter, training, testing, best_params)

print(june_test.to_latex(index=False))
cf.plot_cap_curve(jn_y_test, jn_y_pred)
```

## Feature Importance

We are using a Linear Kernel to extrac feature importances since SVM rbf kernel does not have a feature_importances_ attribute.

```python
# Linear SVM
best_params = {'C':1.0, 'dual': False, 'penalty': 'l2', 'tol': 1e-5}
               
june_test, jn_y_test, jn_y_pred, jn_x_test , jn_df, jn_classifier, ct, jn_x_train, jn_y_train = cf.run_linearsvm_combine(months, data_path, filter, training, testing, best_params)

print(june_test.to_latex(index=False))
june_features = cf.feature_importance_lr(jn_df, jn_classifier, ct)
cf.plot_cap_curve(jn_y_test, jn_y_pred)
```
