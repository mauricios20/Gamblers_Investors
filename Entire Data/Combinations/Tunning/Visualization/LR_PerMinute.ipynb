```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import counting_fns as cf
## Logistic Regression
# Months
months = ['June', 'July', 'August', 'September', 'October']

# Set the path to the csv files
data_path = '/Users/mau/Library/CloudStorage/Dropbox/Mac/Documents/Dissertation/Chapter 2/Entire_Data/Feauture Importance/Minutes/LR/'

times = ["1min", "5min", "10min", "15min"]

# Create a list of dataframes
datasets = {}

# Read the csv files and append them to the list
for month in months:
    for time in times:
        name = month + '_' + time
        data = pd.read_csv(data_path + month + '/coefficients_df_' + time + '.csv')
        datasets[name] = data  

# Let's concact all 1min dataframes 
df_1min = pd.concat([datasets['June_1min'], datasets['July_1min'], datasets['August_1min'], datasets['September_1min'], datasets['October_1min']])

# Let's concact all 5min dataframes
df_5min = pd.concat([datasets['June_5min'], datasets['July_5min'], datasets['August_5min'], datasets['September_5min'], datasets['October_5min']])

# Let's concact all 10min dataframes
df_10min = pd.concat([datasets['June_10min'], datasets['July_10min'], datasets['August_10min'], datasets['September_10min'], datasets['October_10min']])

# Let's concact all 15min dataframes
df_15min = pd.concat([datasets['June_15min'], datasets['July_15min'], datasets['August_15min'], datasets['September_15min'], datasets['October_15min']])

# Drop the first column 
df_1min = df_1min.drop(df_1min.columns[0], axis=1)
df_5min = df_5min.drop(df_5min.columns[0], axis=1)
df_10min = df_10min.drop(df_10min.columns[0], axis=1)
df_15min = df_15min.drop(df_15min.columns[0], axis=1)
```

```python
## Gruop dataframe to only show top 10 features
output_1min = cf.process_dataframe(df_1min, 'Feature', 'Coefficient', top_n=10)
output_5min = cf.process_dataframe(df_5min, 'Feature', 'Coefficient', top_n=10)
output_10min = cf.process_dataframe(df_10min, 'Feature', 'Coefficient', top_n=10)
output_15min = cf.process_dataframe(df_15min, 'Feature', 'Coefficient', top_n=10)

# Delete 'encoder__' from the feature names
output_1min['Feature'] = output_1min['Feature'].str.replace('encoder__age_gen_', '')
output_5min['Feature'] = output_5min['Feature'].str.replace('encoder__age_gen_', '')
output_10min['Feature'] = output_10min['Feature'].str.replace('encoder__age_gen_', '')
output_15min['Feature'] = output_15min['Feature'].str.replace('encoder__age_gen_', '')
```
